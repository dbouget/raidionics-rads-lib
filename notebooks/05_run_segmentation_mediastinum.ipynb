```python
#Download the test data
import os
import requests
import zipfile

test_patient_url = 'https://github.com/raidionics/Raidionics-models/releases/download/1.2.0/Samples-RaidionicsRADSLib-UnitTest3.zip'
lungs_model_url = 'https://github.com/raidionics/Raidionics-models/releases/download/1.2.0/Raidionics-CT_Lungs-ONNX-v12.zip'
airways_model_url = 'https://github.com/raidionics/Raidionics-models/releases/download/1.2.0/Raidionics-CT_Airways-ONNX-v12.zip'
lymphnodes_model_url = 'https://github.com/raidionics/Raidionics-models/releases/download/1.2.0/Raidionics-CT_LymphNodes-ONNX-v12.zip'
pulmsyst_model_url = 'https://github.com/raidionics/Raidionics-models/releases/download/1.2.0/Raidionics-CT_PulmSysHeart-ONNX-v12.zip'
mediumorgans_model_url = 'https://github.com/raidionics/Raidionics-models/releases/download/1.2.0/Raidionics-CT_MediumOrgansMediastinum-ONNX-v12.zip'

test_dir = os.path.join(os.getcwd(), 'unit_tests_results_dir')
os.makedirs(test_dir, exist_ok=True)
patient_dir = os.path.join(test_dir, 'patients')
models_dir = os.path.join(test_dir, 'models')
results_dir = os.path.join(test_dir, 'results')
os.makedirs(patient_dir, exist_ok=True)
os.makedirs(models_dir, exist_ok=True)
os.makedirs(results_dir, exist_ok=True)

archive_dl_dest = os.path.join(test_dir, 'test_patient3.zip')
if not os.path.exists(archive_dl_dest):
    headers = {}
    response = requests.get(test_patient_url, headers=headers, stream=True)
    response.raise_for_status()
    if response.status_code == requests.codes.ok:
        with open(archive_dl_dest, "wb") as f:
            for chunk in response.iter_content(chunk_size=1048576):
                f.write(chunk)
    with zipfile.ZipFile(archive_dl_dest, 'r') as zip_ref:
        zip_ref.extractall(patient_dir)
           
archive_dl_dest = os.path.join(test_dir, 'lungs-model.zip')
if not os.path.exists(archive_dl_dest):
    headers = {}
    response = requests.get(lungs_model_url, headers=headers, stream=True)
    response.raise_for_status()
    if response.status_code == requests.codes.ok:
        with open(archive_dl_dest, "wb") as f:
            for chunk in response.iter_content(chunk_size=1048576):
                f.write(chunk)
    with zipfile.ZipFile(archive_dl_dest, 'r') as zip_ref:
        zip_ref.extractall(models_dir)
    
archive_dl_dest = os.path.join(test_dir, 'airways-model.zip')
if not os.path.exists(archive_dl_dest):
    headers = {}
    response = requests.get(airways_model_url, headers=headers, stream=True)
    response.raise_for_status()
    if response.status_code == requests.codes.ok:
        with open(archive_dl_dest, "wb") as f:
            for chunk in response.iter_content(chunk_size=1048576):
                f.write(chunk)
    with zipfile.ZipFile(archive_dl_dest, 'r') as zip_ref:
        zip_ref.extractall(models_dir)

archive_dl_dest = os.path.join(test_dir, 'lymphnodes-model.zip')
if not os.path.exists(archive_dl_dest):
    headers = {}
    response = requests.get(lymphnodes_model_url, headers=headers, stream=True)
    response.raise_for_status()
    if response.status_code == requests.codes.ok:
        with open(archive_dl_dest, "wb") as f:
            for chunk in response.iter_content(chunk_size=1048576):
                f.write(chunk)
    with zipfile.ZipFile(archive_dl_dest, 'r') as zip_ref:
        zip_ref.extractall(models_dir)
        
archive_dl_dest = os.path.join(test_dir, 'pulmsyst-model.zip')
if not os.path.exists(archive_dl_dest):
    headers = {}
    response = requests.get(pulmsyst_model_url, headers=headers, stream=True)
    response.raise_for_status()
    if response.status_code == requests.codes.ok:
        with open(archive_dl_dest, "wb") as f:
            for chunk in response.iter_content(chunk_size=1048576):
                f.write(chunk)
    with zipfile.ZipFile(archive_dl_dest, 'r') as zip_ref:
        zip_ref.extractall(models_dir)

archive_dl_dest = os.path.join(test_dir, 'mediumorgans-model.zip')
if not os.path.exists(archive_dl_dest):
    headers = {}
    response = requests.get(mediumorgans_model_url, headers=headers, stream=True)
    response.raise_for_status()
    if response.status_code == requests.codes.ok:
        with open(archive_dl_dest, "wb") as f:
            for chunk in response.iter_content(chunk_size=1048576):
                f.write(chunk)
    with zipfile.ZipFile(archive_dl_dest, 'r') as zip_ref:
        zip_ref.extractall(models_dir)
        
assert(os.path.exists(os.path.join(models_dir, 'CT_Lungs')))
assert(os.path.exists(os.path.join(models_dir, 'CT_Airways')))
assert(os.path.exists(os.path.join(models_dir, 'CT_LymphNodes')))
assert(os.path.exists(os.path.join(models_dir, 'CT_PulmSystHeart')))
assert(os.path.exists(os.path.join(models_dir, 'CT_MediumOrgansMediastinum')))
assert(os.path.exists(os.path.join(patient_dir, 'patient-UnitTest3')))
```

```python
# Prepare the configuration file
import configparser
import json
import logging

logging.basicConfig()
logging.getLogger().setLevel(logging.DEBUG)

using_lungs_stripped_inputs = False
using_co_registered_inputs = False

rads_config = configparser.ConfigParser()
rads_config.add_section('Default')
rads_config.set('Default', 'task', 'mediastinum_diagnosis')
rads_config.set('Default', 'caller', '')
rads_config.add_section('System')
rads_config.set('System', 'gpu_id', "-1")
rads_config.set('System', 'input_folder', os.path.join(patient_dir, 'patient-UnitTest3'))
rads_config.set('System', 'output_folder', results_dir)
rads_config.set('System', 'model_folder', models_dir)
rads_config.set('System', 'pipeline_filename', os.path.join(test_dir, 'test_pipeline.json'))
rads_config.add_section('Runtime')
rads_config.set('Runtime', 'reconstruction_method', 'probabilities')
rads_config.set('Runtime', 'reconstruction_order', 'resample_first')
rads_config.set('Runtime', 'use_stripped_data', 'True' if using_lungs_stripped_inputs else 'False')
rads_config.set('Runtime', 'use_registered_data', 'True' if using_co_registered_inputs else 'False')
rads_config_filename = os.path.join(results_dir, 'rads_config.ini')
with open(rads_config_filename, 'w') as outfile:
    rads_config.write(outfile)
    
# Prepare the underlying pipeline
pipeline_json = {}
step_index = 1
step_str = str(step_index)
if not using_lungs_stripped_inputs:
    pipeline_json[step_str] = {}
    pipeline_json[step_str]["task"] = "Segmentation"
    pipeline_json[step_str]["inputs"] = {}
    pipeline_json[step_str]["inputs"]["0"] = {}
    pipeline_json[step_str]["inputs"]["0"]["timestamp"] = 0
    pipeline_json[step_str]["inputs"]["0"]["sequence"] = "High-resolution"
    pipeline_json[step_str]["inputs"]["0"]["labels"] = None
    pipeline_json[step_str]["inputs"]["0"]["space"] = {}
    pipeline_json[step_str]["inputs"]["0"]["space"]["timestamp"] = 0
    pipeline_json[step_str]["inputs"]["0"]["space"]["sequence"] = "High-resolution"
    pipeline_json[step_str]["target"] = "Lungs"
    pipeline_json[step_str]["model"] = "CT_Lungs"
    pipeline_json[step_str]["description"] = "Lungs segmentation in High-resolution (T0)."
    step_index = step_index + 1
    
step_str = str(step_index)
pipeline_json[step_str] = {}
pipeline_json[step_str]["task"] = "Segmentation"
pipeline_json[step_str]["inputs"] = {}
pipeline_json[step_str]["inputs"]["0"] = {}
pipeline_json[step_str]["inputs"]["0"]["timestamp"] = 0
pipeline_json[step_str]["inputs"]["0"]["sequence"] = "High-resolution"
pipeline_json[step_str]["inputs"]["0"]["labels"] = None
pipeline_json[step_str]["inputs"]["0"]["space"] = {}
pipeline_json[step_str]["inputs"]["0"]["space"]["timestamp"] = 0
pipeline_json[step_str]["inputs"]["0"]["space"]["sequence"] = "High-resolution"
pipeline_json[step_str]["target"] = "Airways"
pipeline_json[step_str]["model"] = "CT_Airways"
pipeline_json[step_str]["description"] = "Airways segmentation in High-resolution (T0)."

step_index = step_index + 1
step_str = str(step_index)
pipeline_json[step_str] = {}
pipeline_json[step_str]["task"] = "Segmentation"
pipeline_json[step_str]["inputs"] = {}
pipeline_json[step_str]["inputs"]["0"] = {}
pipeline_json[step_str]["inputs"]["0"]["timestamp"] = 0
pipeline_json[step_str]["inputs"]["0"]["sequence"] = "High-resolution"
pipeline_json[step_str]["inputs"]["0"]["labels"] = None
pipeline_json[step_str]["inputs"]["0"]["space"] = {}
pipeline_json[step_str]["inputs"]["0"]["space"]["timestamp"] = 0
pipeline_json[step_str]["inputs"]["0"]["space"]["sequence"] = "High-resolution"
pipeline_json[step_str]["target"] = "LymphNodes"
pipeline_json[step_str]["model"] = "CT_LymphNodes"
pipeline_json[step_str]["description"] = "Lymph nodes segmentation in High-resolution (T0)."

step_index = step_index + 1
step_str = str(step_index)
pipeline_json[step_str] = {}
pipeline_json[step_str]["task"] = "Segmentation"
pipeline_json[step_str]["inputs"] = {}
pipeline_json[step_str]["inputs"]["0"] = {}
pipeline_json[step_str]["inputs"]["0"]["timestamp"] = 0
pipeline_json[step_str]["inputs"]["0"]["sequence"] = "High-resolution"
pipeline_json[step_str]["inputs"]["0"]["labels"] = None
pipeline_json[step_str]["inputs"]["0"]["space"] = {}
pipeline_json[step_str]["inputs"]["0"]["space"]["timestamp"] = 0
pipeline_json[step_str]["inputs"]["0"]["space"]["sequence"] = "High-resolution"
pipeline_json[step_str]["target"] = "Heart", "PulmonaryVeins", "PulmonaryTrunk"
pipeline_json[step_str]["model"] = "CT_PulmSystHeart"
pipeline_json[step_str]["description"] = "Heart, pulmonary veins and trunk segmentation in High-resolution (T0)."

step_index = step_index + 1
step_str = str(step_index)
pipeline_json[step_str] = {}
pipeline_json[step_str]["task"] = "Segmentation"
pipeline_json[step_str]["inputs"] = {}
pipeline_json[step_str]["inputs"]["0"] = {}
pipeline_json[step_str]["inputs"]["0"]["timestamp"] = 0
pipeline_json[step_str]["inputs"]["0"]["sequence"] = "High-resolution"
pipeline_json[step_str]["inputs"]["0"]["labels"] = None
pipeline_json[step_str]["inputs"]["0"]["space"] = {}
pipeline_json[step_str]["inputs"]["0"]["space"]["timestamp"] = 0
pipeline_json[step_str]["inputs"]["0"]["space"]["sequence"] = "High-resolution"
pipeline_json[step_str]["target"] ="VenaCava", "AorticArch", "AscendingAorta", "DescendingAorta", "Spine"
pipeline_json[step_str]["model"] = "CT_MediumOrgansMediastinum"
pipeline_json[step_str]["description"] = "Vena cava, aortic arch, ascending aorta, descending aorta, and spine segmentation in High-resolution (T0)."

with open(os.path.join(test_dir, 'test_pipeline.json'), 'w', newline='\n') as outfile:
    json.dump(pipeline_json, outfile, indent=4)

# Run the segmentation task
from raidionicsrads.compute import run_rads
run_rads(rads_config_filename)
```

```python
import nibabel as nib
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

logging.getLogger().setLevel(logging.WARNING)
data_filename = os.path.join(patient_dir, 'patient-UnitTest3', 'T0', '1_CT_HR.nii.gz')
lungs_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-Lungs.nii.gz')
airways_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-Airways.nii.gz')
lymphnodes_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-LymphNodes.nii.gz')
heart_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-Heart.nii.gz')
pulmtrunk_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-PulmonaryTrunk.nii.gz')
pulmveins_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-PulmonaryVeins.nii.gz')
spine_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-Spine.nii.gz')
venacava_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-VenaCava.nii.gz')
ascaorta_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-AscendingAorta.nii.gz')
descaorta_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-DescendingAorta.nii.gz')
aortarch_predictions_filename = os.path.join(results_dir, 'T0', '1_CT_HR_annotation-AorticArch.nii.gz')

data = nib.load(data_filename).get_fdata()[:]
lungs_predictions = nib.load(lungs_predictions_filename).get_fdata()[:]
airways_predictions = nib.load(airways_predictions_filename).get_fdata()[:]
lymphnodes_predictions = nib.load(lymphnodes_predictions_filename).get_fdata()[:]
heart_predictions = nib.load(heart_predictions_filename).get_fdata()[:]
pulmtrunk_predictions = nib.load(pulmtrunk_predictions_filename).get_fdata()[:]
pulmveins_predictions = nib.load(pulmveins_predictions_filename).get_fdata()[:]
spine_predictions = nib.load(spine_predictions_filename).get_fdata()[:]
venacava_predictions = nib.load(venacava_predictions_filename).get_fdata()[:]
ascaorta_predictions = nib.load(ascaorta_predictions_filename).get_fdata()[:]
descaorta_predictions = nib.load(descaorta_predictions_filename).get_fdata()[:]
aortarch_predictions = nib.load(aortarch_predictions_filename).get_fdata()[:]

all_predictions = np.zeros(data.shape).astype('uint8')
all_predictions[lungs_predictions >= 0.5] = 1
all_predictions[airways_predictions >= 0.5] = 2
all_predictions[lymphnodes_predictions >= 0.5] = 3
all_predictions[heart_predictions >= 0.5] = 4
all_predictions[pulmtrunk_predictions >= 0.5] = 5
all_predictions[pulmveins_predictions >= 0.5] = 6
all_predictions[venacava_predictions >= 0.5] = 7
all_predictions[spine_predictions >= 0.5] = 8
all_predictions[ascaorta_predictions >= 0.5] = 9
all_predictions[descaorta_predictions >= 0.5] = 9
all_predictions[aortarch_predictions >= 0.5] = 9

# Plotting corresponding inputs and predictions for a single 2D slice
colors = ['black','green','blue','red', 'purple', 'orange', 'brown', 'pink', 'olive', 'cyan']
cmap = matplotlib.colors.ListedColormap(colors)

slice_nr = 472
fig, ax = plt.subplots(3, 3, figsize=(20, 20))
ax[0, 0].imshow(np.rot90(data[..., slice_nr], k=3), cmap="gray")
ax[0, 1].imshow(np.rot90(all_predictions[..., slice_nr], k=3), cmap=cmap)
ax[0, 2].imshow(np.rot90(lungs_predictions[..., slice_nr], k=3), cmap="Reds")
ax[1, 0].imshow(np.rot90(airways_predictions[..., slice_nr], k=3), cmap="Reds")
ax[1, 1].imshow(np.rot90(pulmtrunk_predictions[..., slice_nr], k=3), cmap="Reds")
ax[1, 2].imshow(np.rot90(pulmveins_predictions[..., slice_nr], k=3), cmap="Reds")
ax[2, 0].imshow(np.rot90(lymphnodes_predictions[..., slice_nr], k=3), cmap="Reds")
ax[2, 1].imshow(np.rot90(venacava_predictions[..., slice_nr], k=3), cmap="Reds")
ax[2, 2].imshow(np.rot90(spine_predictions[..., slice_nr], k=3), cmap="Reds")


for i, title in enumerate(["Data", "All labels", "Lungs", "Airways", "Pulmonary trunk", "Pulmonary veins", "Lymph Nodes", "Vena cava", "Spine"]):
  ax[int(i/3), i%3].set_title(title)
plt.show()
```
